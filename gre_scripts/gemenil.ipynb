```python
def wordlist_to_string(group):
    # create a string of words separated by newlines
    # the string should be in the format of:
    group = group
    words_string = ''
    for word in group:
        # add the word to the words_string
        words_string += '\n'
        words_string += word
    return words_string
wordlist_to_string(words_by_group[19])
```

```python
import base64
import os
from threading import local
# Update imports to use the correct package
import google.generativeai as genai
from google.generativeai import types
import time


def generate_conversation(words):
    Words = words
    # Configure the API key
    genai.configure(api_key=os.environ.get("GEMINI_API_KEY"))

    # Create a chat session
    model = genai.GenerativeModel("gemini-2.0-flash-thinking-exp-01-21")
    chat = model.start_chat(history=[])
    
    # First prompt - detailed definitions and example sentences
    first_prompt = f"""For the following GRE vocabulary words, provide:
1. A clear and concise definition
2. Two sentences that demonstrate the word's proper use; make the sentences concise keeping the focus on the portraying the meaning of the word so that the reader can grasp the new word better.

Words:
{Words}
"""
    
    # Send first prompt and get response
    first_response = chat.send_message(first_prompt)
    time.sleep(10)
    
    # Second prompt - concise definitions
    second_prompt = """Now give me meaning of the words in one words, if not possible in one word you can use up to a phrase five words, no more"""
    
    # Send second prompt and get response
    second_response = chat.send_message(second_prompt)
    time.sleep(10)


    third_prompt = """Make a google sheet with them in the format:
Word	concise_meaning sentence_1	full_meaning	sentence_2"""
    
    # Send second prompt and get response
    third_response = chat.send_message(third_prompt)
    time.sleep(10)
    
    return chat
```
